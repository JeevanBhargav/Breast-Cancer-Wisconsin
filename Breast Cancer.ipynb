```python
bc['Bare Nuclei'] = pd.to_numeric(bc['Bare Nuclei'], errors='coerce')
```

```python
bc['Bare Nuclei'].fillna(bc['Bare Nuclei'].mean(), inplace = True)
```

```python
for i in range (0,11):
    bc[col_Names[i]] = pd.to_numeric(bc[col_Names[i]], errors='coerce')
    print(bc[col_Names[i]].isna().sum())
```

```python
bc.corr()
```

##### As we can see, that the correlation between target variable and sample code number is negative. So in the next step we are dropping sample code number from the dataframe.

```python
bc.drop("Sample code number",axis=1,inplace=True) 
```

```python
col_Names=["Clump thickness", "Uniformity of cell Size", "Uniformity of cell shape", "marginal Ahesion","Single epithelial cell size", "Bare Nuclei", "Bland Chromatin", "Normal Nucleoli", "Mitoses"]
```

```python
for i in col_Names:
    
    bc[(bc[col_Names] > bc[col_Names].mean() + 3*bc[col_Names].std()) | (bc[col_Names] < bc[col_Names].mean() - 3*bc[col_Names].std())]
```

```python
for i in col_Names:
    
    upper_limit = bc[col_Names].mean() + 3*bc[col_Names].std()
    lower_limit = bc[col_Names].mean() - 3*bc[col_Names].std()
    bc[col_Names] = np.where(
        bc[col_Names]>upper_limit,
        upper_limit,
        np.where(
            bc[col_Names]<lower_limit,
            lower_limit,
            bc[col_Names]
        )
    )
```

```python
X = bc.drop(columns= 'Class')
y = bc.Class
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, stratify = y, random_state = 1)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=1)

dt.fit(X_train, y_train)

pred = dt.predict(X_test)

dt_score = dt.score(X_test, y_test)

result = pd.DataFrame({'Model' : ['Decision Tree Classifier'], 'Accuracy' : [dt_score]})

result
```

## Decision Tree with K fold 

```python
from sklearn.model_selection import cross_val_score
scores= cross_val_score(dt,X,y,cv=10,scoring='accuracy')
```

```python
scores
```

```python
scores.mean()
```

#### We observe that there is 0.001 increase in accuracy when used k-fold cross validation

## KNN 

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
```

```python
 knn = KNeighborsClassifier(n_neighbors= 10)
```

```python
scores_knn= cross_val_score(knn,X,y,cv=10,scoring='accuracy')
```

```python
scores_knn
```

```python
scores_knn.mean()
```

```python
x=0
k=0
for i in range (1,51):
    knn=KNeighborsClassifier(n_neighbors= i)
    
    scores_knn= cross_val_score(knn,X,y,cv=10,scoring='accuracy')
    
    scores_knn.mean()
    
    print("For k = "+ str(i) +"- Acccuracy = "+str(scores_knn.mean()))
    
    if(scores_knn.mean()>x):
        x=scores_knn.mean()
        k=i

print("The best accuracy levels are achieved when k is "+ str(k))
```

## SVM 

```python
from sklearn.svm import SVC
```

```python
j=5
svm_kernels=["linear","sigmoid","poly"]
for i in svm_kernels:
    svm_classifier = SVC(kernel=i,random_state=1)
    
    cv_scores_1 = cross_val_score(svm_classifier,X,y,scoring='accuracy',cv=10)
    print("When kernel = "+i)
    print(cv_scores_1)
    print(np.mean(cv_scores_1))
    
    result.loc[j] = ['svm.SVC when kernel = '+i, np.mean(cv_scores_1)]
    j+=1
```

##### When the Kernel is Linear, we got the highest accuracy of 96.4%
